```python
import numpy as np
from sklearn.decomposition import PCA
from pandas import *
import datetime as dt
import statsmodels.api as sm
```

```python
prices_long = read_csv('data_set/stock_prices.csv', index_col=[0,1], parse_dates=True, squeeze=True)
dji_all = read_csv('data_set/DJI.csv', index_col=0, parse_dates=True, squeeze=True)
print prices_long.head(5)
```

```python
calc_ret = lambda x: np.log(x / x.shift(1))[1:]
scale = lambda x: (x - x.mean()) / x.std()
# scale = lambda x: x - x.mean()

def OLSreg(y, Xmat):
    return sm.OLS(y, sm.add_constant(Xmat, prepend=True)).fit()
```

```python
prices = prices_long.unstack()
print prices_long.head(2)
dates = list(prices.index)
dates.remove(dt.datetime(2002,2,1))
prices = prices.drop('DDR', axis=1)
prices = prices.drop([dt.datetime(2002,2,1)])
```

```python
print len(dates)
```

```python
print type(prices)
returns = prices.apply(calc_ret)
```

```python
dji_all = dji_all.sort_index()
dji = dji_all['Close'].reindex(index = dates)
print dji.head(5)
dji_ret = calc_ret(dji)
print dji_ret.head(5)
```

```python
def make_pca_index(data, scale_data = True):
    if scale_data:
        data_std = data.apply(scale)
    else:
        data_std = data
    corrs = data_std.corr()
    pca = PCA(n_components=1).fit(corrs)
    mkt_index = -scale(pca.transform(data_std))
    return mkt_index
```

```python
def make_pca_index2(data, scale_data = True):
    if scale_data:
        data_std = data.apply(scale)
    else:
        data_std = data
    pca = PCA(n_components=1).fit(data_std)
    mkt_index = scale(pca.transform(data_std))
    return mkt_index
```

```python
price_index = make_pca_index2(prices)
```

```python
plt.figure(figsize=(17, 5))
plt.subplot(121)
plt.plot(price_index, scale(dji), 'k.')
ols_fit = OLSreg(scale(dji), price_index)
plt.plot(price_index, ols_fit.fittedvalues, 'r-')

plt.subplot(122)
plt.plot(dates, price_index, label='PCA Price Index')
plt.plot(dates, scale(dji), label='DJ Index')
plt.legend(loc = 'upper left')
```

```python
returns_index = make_pca_index(returns)
print type(returns)
```

```python
ols_fit = OLSreg(dji_ret, returns_index)
plt.figure(figsize=(17,5))
plt.subplot(121)
plt.plot(returns_index, dji_ret, 'k.')
plt.plot(returns_index, ols_fit.fittedvalues, 'r-')

plt.subplot(122)
plt.plot(dates[1:], np.exp((returns_index * dji_ret.std()).cumsum()), label='PCA Return')
plt.plot(dates[1:], np.exp(dji_ret.cumsum()), label='Index')
plt.legend(loc = 'upper left')
print dji_ret.std()
```
