```python
import pandas as pd
import json
import seaborn
%matplotlib inline
```

```python
db = json.load(open('data_set/foods-2011-10-03.json'))
len(db)
```

```python
db[0].keys()
```

```python
print db[0]['nutrients'][0]
nutrients = pd.DataFrame(db[0]['nutrients'])
nutrients[:7]
```

```python
info_keys = ['description', 'group', 'id', 'manufaturer']
info = pd.DataFrame(db, columns=info_keys)
info[:5]
```

```python
info.group.value_counts()
```

```python
nutrients = []
for rec in db:
    fnuts = pd.DataFrame(rec['nutrients'])
    fnuts['id'] = rec['id']
    nutrients.append(fnuts)
nutrients = pd.concat(nutrients, ignore_index=True)
```

```python
nutrients.head()
```

```python
nutrients = nutrients.drop_duplicates()
```

```python
info = info.rename(columns={'description': 'food', 'group': 'fgroup'})
```

```python
nutrients = nutrients.rename(columns={'description': 'nutrient', 'group': 'nutgroup'})
```

```python
nutrients.head()
```

```python
info.head()
```

```python
ndata = pd.merge(nutrients, info, on='id', how='outer')
ndata.head()
```

```python
result = ndata.groupby(['nutrient', 'fgroup'])['value'].quantile(0.5)
result['Zinc, Zn'].order().plot(kind='barh')
```
