# Regression Plot

Learn From StatsModels 
http://statsmodels.sourceforge.net/devel/examples/notebooks/generated/example_regression_plots.html

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn
%pylab inline
```

```python
prestige = sm.datasets.get_rdataset('Duncan', 'car', cache=True).data
prestige.head()
```

```python
prestige_model = ols('prestige ~ income + education', data =prestige).fit()
prestige_model.summary()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
fig = sm.graphics.influence_plot(prestige_model, ax = ax, criterion='cooks')
```

### Partial Regression Plots

```python
fig, ax = plt.subplots(figsize = (12, 5))
fig = sm.graphics.plot_partregress('prestige', 'income', ['education' ,'income'], data=prestige, ax=ax, obs_labels=True)
ax = fig.axes[0]
```

```python
subset = ~prestige.index.isin(['conductor', 'RR.engineer', 'minister'])
prestige_model2 = ols('prestige ~ income + education', data = prestige, subset=subset).fit()
prestige_model2.summary()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
fig = sm.graphics.influence_plot(prestige_model2, ax=ax, criterion='cook')
```

```python
fig, ax = plt.subplots(figsize=(10,8))
fig = sm.graphics.plot_partregress_grid(prestige_model2, fig=fig)
```

### Example: Statewide Crime 2009 Dataset

```python
data = sm.datasets.statecrime.load_pandas().data
data.head()
```

```python
crime_model = ols('murder ~ violent + poverty + urban + hs_grad + single', data = data).fit()
crime_model.summary()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
fig = sm.graphics.plot_partregress_grid(crime_model, fig=fig)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
fig = sm.graphics.plot_partregress('murder', 'hs_grad', ['violent', 'poverty', 'urban', 'single'], data = data, ax=ax)
```

### Leverage-Resid2 Plot

```python
fig, ax = plt.subplots(figsize=(8,6))
fig = sm.graphics.plot_leverage_resid2(crime_model, ax=ax)
```

### Influence Plot

```python
fig, ax = plt.subplots(figsize=(8,6))
fig = sm.graphics.influence_plot(crime_model, ax=ax)
```
