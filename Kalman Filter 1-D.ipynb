```python
import matplotlib.pyplot as plt
import numpy.random as random
import math
import numpy as np
from pg_dao import financial_dao
```

```python
class KalmanFilter1D:
    def __init__(self, x0, P, R, Q):
        self.x = x0
        self.P = P
        self.R = R
        self.Q = Q
        
    def update(self, z):
        self.x = (self.P * z + self.x * self.R) / (self.P + self.R)
        self.P = 1. / (1./self.P + 1./self.R)


    def predict(self, u=0.0):
        self.x += u
        self.P += self.Q
```

```python
data = financial_dao.get_price_from_date('tw', '5457.TWO', '2012-01-01')
nav = [nav for da, ticker, nav in data]
variance = np.var(nav)
monthly_nav = [nav[int(x)] for x in np.linspace(0, len(nav)-2, 20)]
movement_variance = np.var(monthly_nav)/ math.sqrt(20.0)
print variance, movement_variance
estimates = [nav[1]]
kf = KalmanFilter1D(x0=nav[1], 
                    P=10000, 
                    R=variance, 
                    Q=movement_variance)
for i in range(len(nav)):
    kf.predict()
    estimates.append(kf.x)
    kf.update(nav[i])
#     estimates.append(kf.x)
plt.figure(figsize=(15,8))
plt.plot(nav, 'b', label='Equity')
plt.plot(estimates, 'r-', label='Filter')
plt.legend(loc='best')
```

```python
def sma(data, window):
    result=[]
    temp = 0
    for i in range(window):
        result.append(data[i])
        temp += data[i]
    for i in range(window, len(data)):
        temp = temp + data[i] - data[i-window]
        result.append(temp/window)
    return result
def predict(data):
    result=[data[0], data[1]]
    for i in range(2,len(data)-1):
        result.append(2*data[i-1]-data[i-2])
    return result
```

```python
ma = sma(nav, 10)
ma_predict = predict(ma)
plt.figure(figsize = (15,10))
plt.plot(nav, label='nav')
plt.plot(ma_predict, label='predict')
plt.legend(loc='best')
```

```python
err_ma = []
for i in range(1,len(nav)-1):
    err_ma.append((nav[i]-ma_predict[i])/nav[i])
df_ma = pd.DataFrame(err_ma, columns=['ma'])
df_ma.hist(bins=50)
```
