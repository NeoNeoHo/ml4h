```python
from __future__ import print_function, division
import matplotlib.pyplot as plt
import numpy.random as random
import math
import numpy as np
import seaborn as sns

class DogSensor(object):
    
    def __init__(self, x0=0, velocity=1, variance=0.0):
        self.x = x0
        self.velocity = velocity
        self.noise = math.sqrt(variance)
        
    def sense_position(self):
        self.x = self.x + self.velocity
        return self.x + random.randn() * self.noise
```

```python
def test_sensor(variance):
    dog = DogSensor(variance=variance)
    
    xs = []
    for i in range(100):
        x = dog.sense_position()
        xs.append(x)
    
    plt.figure(figsize=(10,5))
    plt.plot(xs, label='sensor')
    plt.plot([0,99], [1,100], 'r--', label='actual')
    plt.xlabel('time')
    plt.ylabel('pos')
    plt.title('variance = ' + str(variance))
    plt.legend(loc='best')

test_sensor(variance=16)
```

```python
def multiply(mu1, var1, mu2, var2):
    if var1 == 0.0:
        var1=1.e-80
        
    if var2 == 0:
        var2 = 1e-80
        
    mean = (var1*mu2 + var2*mu1) / (var1+var2)
    variance = 1 / (1/var1 + 1/var2)
    return (mean, variance)
```

```python
def update(mean, variance, measurement, measurement_variance):
    return multiply(mean, variance, measurement, measurement_variance)
```

```python
dog = DogSensor(velocity=0., variance=1.)
pos, s = 2, 5
for i in range(20):
    pos, s = update(pos, s, dog.sense_position(), 5)
    print('time:', i,
          '\t position = ', '%.3f' %pos, 
          '\t variance = ', '%.3f' %s)
```

```python
def predict(pos, variance, movement, movement_variance):
    return (pos + movement, variance + movement_variance)
```

```python
movement = 1
movement_variance = 2
sensor_variance = 10
pos = (0, 500)

dog = DogSensor(pos[0], velocity=movement, variance=sensor_variance)

zs = []
ps = []
pps = []

for i in range(50):
    pos = predict(pos[0], pos[1], movement, movement_variance)
#     print('Predict: {: 10.4f}{: 10.4f}'.format(pos[0], pos[1]), end='\t')
    pps.append(pos[0])
    Z = dog.sense_position()
    zs.append(Z)
    
    pos = update(pos[0], pos[1], Z, sensor_variance)
    ps.append(pos[0])
    
#     print('Update: {: 10.4f}{: 10.4f}'.format(pos[0], pos[1]))
plt.figure(figsize=(10,4))
plt.plot(ps, label='filter')
plt.plot(pps, label = 'predict')
plt.plot(zs, c='r', linestyle='dashed', label='measurement')
plt.legend(loc='best')
plt.show()
```

```python
class KalmanFilter1D:
    def __init__(self, x0, P, R, Q):
        self.x = x0    # state
        self.P = P     # variance of the state
        self.R = R     # variance of the measurement
        self.Q = Q   # variance of the movement
    
    def update(self, z):
        self.x = (self.P * z + self.x * self.R) / (self.P + self.R)
        self.P = 1. / (1./self.P + 1./self.R)
    
    def predict(self, u=0.0):
        self.x += u
        self.P += self.Q
```

```python
def volt(voltage, temp_variance):
    return random.randn()*temp_variance + voltage
```

```python
variance = 300
movement_variance =2
actual_voltage = 0
movement = 1

N = 50
zs = [volt((i-25)**2, variance) for i in range(N)]
ps = []
estimates = []

kf = KalmanFilter1D(x0=0,   # initial state
                    P = 500,            # initial variance
                    R = variance,      # sensor noise
                    Q = movement_variance)  # movement noise

for i in range(N):
    kf.predict(u = movement)
    kf.update(zs[i])
    
    estimates.append(kf.x)
    ps.append(kf.P)
    
# plot the filter output and the variance
plt.plot(zs, c='r', linestyle='dashed',label='measurements')
plt.plot(estimates, label='filter')
plt.legend(loc='best')
```
